Supuestos del modelo:

- Uniformidad: todas las empresas del grupo son aproximadamente idénticas, por lo que enfrentan condiciones de costos y demanda similares
- Las empresas se enfrentan a una curva de demanda negativa
- Productos heterogéneos, pero sustitutos cercanos

Debido a la diferenciación de productos, tendrán un comportamiento similar al del monopolio


## Problema de la Empresa individual

Supongamos que cada empresa individual se enfrenta a la misma función de costos
- CT = a + b * qi para todo bien i entre range(1,n) 
- *n* es la cantidad de productos o marcas (sustitutos cercanos) en el mercado
cada marca es monopolio de si misma pero hay bienes sustitutos cercanos 

Cada empresa, por su parte, se enfrenta a la siguiente función de demanda
- qi = f(m; pi;pj)  para todo i!= j y otros bien(es) j  entre range(1,n-1)
Por tanto, la demanda depende del ingreso de los consumidores, de los precios del bien i y del precio de los otros bienes.
Suponiendo una función de demanda lineal:

- qi(pi;pj) = ci + (n-1)* ((dj * pj) - (ei * pi))
dj = variacion en la demanda de i en función del precio de cada uno de sus sustitutos (el impacto de cada pj en la demanda del bien i)
ei = pendiente con que reacciona la demanda ante la variación de su propio precio (pi)

Supongamos las funciones de demanda a las que se enfrenta cada empresa son iguales.
Planteamos la curva inversa de la demanda, la cual es:
- pi(pi;pj) = fi + (n-1)*(gj * qj) - (hi * qi) ¿la sumatoria es de gj*qj solo?
- gj = como afecta la cantidad demandada de los otros bienes al precio del bien i
- hi = efecto que tiene qi sobre pi

Dado que g es relativamente pequeño y considerando que la acción individual de cada empresa afecta por igual a los demás (n-1) competidores, pueden despreciarse los efectos sobre el precio.
Por tanto, cada empresa se comportará como si su acción no afectase a los demás competidores.

El beneficio del empresario viene dado por:
- beneficio = ingreso - CT = pi(pi;pj) * qi - CT(qi) =>
- beneficio = ingreso - CT = (fi + (n-1)*(gj * qj)) * qi - CT(qi)

Las Condiciones de Primer Orden:
- Beneficio(qi)'qi = IMg - CMg = 0 =>
- Beneficio(qi)'qi = fi + (n-1)*(gj * qj) - (2 * hi * qi) - CMg(qi) = 0

Por tanto, obtenemos que el ingreso marginal se iguala al costo marginal en 
- IMg = CMg
- fi + (n-1)*(gj * qj) - (2 * hi * qi) = CMg(qi) 

La empresa representativa no se moverá a lo largo de la curva de demanda efectiva (dd), sino que lo hará sobre la curva de demanda construida bajo el supuesto de que sus competidores no modifican su output (DD).

Su curva de demanda efectiva se forma suponiendo que las empresas poseen una cuota de mercado constante e igual del mercado, por lo que el precio que cobran es el mismo:

- qi = qj =>
- pi = f - (h + (n+1) * g) * qi   #curva dd



### Equilibrio a corto plazo
La curva de demanda efectiva que representa movimientos idénticos y simultáneos de todos los consumidores (dd) es más inclinada que aquella considerada por la empresa representativa (DD).
La DD muestra la curva de demanda producto de la variación de su nivel de producción.
A medida que todas las empresas varían su output, la forma y posición de dd, que es sólo función de qi, permanece inalterada, y DD, cuya posición depende de los niveles de output de las otras empresas se desliza a lo largo de dd, cortándola siempre al nivel de producción de la empresa representativa.

### Equilibrio a largo plazo
El precio se iguala al CMe como en el caso de la empresa competitiva, es decir, los beneficios son normales u ordinarios (iguales a cero), pero no es igual al CMg. El hecho de no alcanzar el mínimo del CMeT significa que las empresas no están utilizando toda la capacidad de producción y por ello resultan ineficientes en el sentido de Pareto. A su vez, comparando los resultados de ambos tipos de competencia, estas empresas producen un menor nivel de producción a un costo medio total mayor.
