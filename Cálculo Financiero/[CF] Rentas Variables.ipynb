```python
import sympy as sp
sp.init_printing()
from sympy import Function,oo, symbols


#Rentas constantes

i, n, C, a_financiera, s_financiera = sp.symbols('i, n, C, a, s')


def valor(momento_de_valuacion, momento_de_pago, periodos, tasa_de_interes):
    n = periodos
    i = tasa_de_interes
    momento_de_valuacion = momento_de_valuacion #F, Final, A o Actual
    momento_de_pago = momento_de_pago #vencido 1 o adelantado 0    
    if momento_de_pago==1:
        #Vencidas
        if momento_de_valuacion=="F" or momento_de_valuacion=="Final":            
            sp.pprint("VF(1;{n};{i}) = C * ((1+{i})^{n} - 1) / {i}".format(i=str(i), n=str(n)))
        elif momento_de_valuacion=="A" or momento_de_valuacion=="Actual":
            if n == oo or n == "inf":
                # Vencidas perpetuas                
                print("VA(1;oo;i) = C / i")
                print("VA(1;oo;{i}) = C / {i}".format(i=str(i)))
            else:
                print("VA(1;n;i) = C * (1 - (1/(1+i))^n) / i")
                print("VA(1;{n};{i}) = C * (1 - (1/(1+{i}))^{n}) / {i}".format(i=str(i), n=str(n)))
    elif momento_de_pago == 0:
        #Adelantadas
        if momento_de_valuacion=="F" or momento_de_valuacion=="Final":            
            print("VF(0;n;i) = C * ((1+i)^n - 1) * (1+i) / i")
            print("VF(0;{n};{i}) = C * ((1+{i})^{n} - 1) * (1+{i}) / {i}".format(i=str(i), n=str(n)))
        elif momento_de_valuacion=="A" or momento_de_valuacion=="Actual":
            print("VA(0;n;i) = C * (1 - (1/(1+i))) * (1+i) / i")
            print("VA(0;{n};{i}) = C * (1 - (1/(1+{i}))) * (1+{i}) / {i}".format(i=str(i), n=str(n)))
    else:
        printf("Vencido debe ser 1 o 0")
    
valor("A", 1, "inf", 0.003)



```
